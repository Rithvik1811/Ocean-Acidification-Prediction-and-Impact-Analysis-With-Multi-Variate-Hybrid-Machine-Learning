```python
import pandas as pd
from PyCO2SYS import CO2SYS

# Constants
total_scale = 1  
K1K2CONSTANTS = 4  
KSO4CONSTANTS = 1  

silicate_column = 'G2silicate'
phosphate_column = 'G2phosphate'

# Function to calculate aragonite saturation
def calculate_aragonite_saturation(row):
    co2_results = CO2SYS(row['G2tco2'], row['G2talk'], 1, 2, 
                         row['G2salinity'], row['G2temperature'], row['G2temperature'], 
                         row['G2pressure'], total_scale,
                         row[silicate_column], row[phosphate_column], total_scale,
                         K1K2CONSTANTS, KSO4CONSTANTS) 
    return co2_results['OmegaARout']  # Access the 'OmegaARout' key

df['Aragonite Saturation'] = df.apply(calculate_aragonite_saturation, axis=1)
```

```python
import ast
def extract_aragonite_value(value):
    if isinstance(value, str):  # Check if it's a string
        try:
            float_list = ast.literal_eval(value)  # Convert string representation of list to a list
            return float_list[0] 
        except (ValueError, SyntaxError): 
            return value  # Handle errors if not in the correct format 
    else:
        return value  # If not a string, return as is

df_aragonite['G2aragonite_sat'] = df_aragonite['G2aragonite_sat'].apply(extract_aragonite_value)
df_aragonite['G2aragonite_sat'] = pd.to_numeric(df_aragonite['G2aragonite_sat'])  # Convert to numeric
```
