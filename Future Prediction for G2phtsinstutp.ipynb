```python
import pandas as pd

# Generating a range of years from 1973 to 2021
all_years = pd.DataFrame({'Year': range(1973, 2022)})

# Merging your dataset with the complete range of years to fill in missing data
df_ph_merged = pd.merge(all_years, df_ph_trend, left_on='Year', right_on='G2year', how='left')

# Create a new DataFrame without the old G2year column
df_ph_merged_new = df_ph_merged.drop(columns=['G2year'])

# Rename the Year column to G2 year
df_ph_merged_new = df_ph_merged_new.rename(columns={'Year': 'G2 year'})

# You may need to handle missing values based on your dataset's specifics
# For example, if your dataset has a 'pH' column, you might want to fill NaN values
# df_ph_merged_new['pH'].fillna(method='ffill', inplace=True)  # Forward fill missing pH values

print(df_ph_merged_new)
```

```python
import pandas as pd
# Define the window size for moving averages
window_size = 3

# Function to fill missing values using moving averages
def fill_missing_with_moving_avg(df):
    for column in df.columns[1:]:
        for i in range(len(df)):
            if pd.isnull(df.at[i, column]):
                start_idx = max(0, i - window_size + 1)
                end_idx = min(len(df), i + window_size)
                valid_values = df.iloc[start_idx:end_idx][column].dropna()
                if len(valid_values) > 0:
                    df.at[i, column] = valid_values.mean()
    return df

# Apply the fill_missing_with_moving_avg function to fill missing values
df_ph_merged_filled = fill_missing_with_moving_avg(df_ph_merged_new)
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split



# Features and Target
features = df_ph_trend.columns[df_ph_trend.columns != 'G2phtsinsitutp'].tolist()
target = 'G2phtsinsitutp'

# Assign features to X and target variable to y
X = df_ph_trend[features]
y = df_ph_trend[target]

# Train the Decision Tree Regressor 
tree_model = DecisionTreeRegressor() 
tree_model.fit(X, y)

# Train the XGBoost Regressor 
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42) 
xgb_model.fit(X, y)

# Define weights for each model
weight_tree = 0.25  # Weight for Decision Tree
weight_xgb = 0.75 # Weight for XGBoost

# Combine predictions using weighted average
def hybrid_predict(X):
    y_pred_tree = tree_model.predict(X)
    y_pred_xgb = xgb_model.predict(X)
    return (weight_tree * y_pred_tree) + (weight_xgb * y_pred_xgb)

# Generate future years from 2022 to 2221 (200 years)
future_years = np.arange(2022, 2222)

# Create DataFrame for future years
future_features = pd.DataFrame(np.repeat(X.values, len(future_years), axis=0), columns=X.columns)
future_features['G2year'] = np.tile(future_years, len(X))

# Make predictions using the hybrid model for future years
predicted_pH_values = hybrid_predict(future_features[features])

# Create DataFrame to store predicted pH values for future years
predicted_df = pd.DataFrame({'G2year': np.tile(future_years, len(X)), 'Predicted_pH': predicted_pH_values})

# Save the predicted pH values to a CSV file
predicted_df.to_csv('predicted_pH_values.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
# Define the classification function
def classify_pH(pH):
    if pH < 7.8:
        return 'High Danger (Ocean Acidification)'
    elif 7.8 <= pH < 8.2:
        return 'Moderate Risk (At Risk)'
    else:
        return 'Safe Zone (Currently)'

# Apply classification function to create a new column
df_ph_trend['Classification'] = df_ph_trend['G2phtsinsitutp'].apply(classify_pH)

# Group the DataFrame by classification and count the occurrences for existing data
existing_classification_counts = df_ph_trend['Classification'].value_counts()

# Plotting the pie chart for existing data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(existing_classification_counts, labels=existing_classification_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of pH Classifications (Existing Data)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

predicted_df['ph_Classification']=predicted_df['Predicted_pH'].apply(classify_pH)

# Group the DataFrame by classification and count the occurrences for future data
future_classification_counts = predicted_df['ph_Classification'].value_counts()

# Plotting the pie chart for future data
plt.subplot(1, 2, 2)
plt.pie(future_classification_counts, labels=future_classification_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of pH Classifications (Future Data)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add legends
plt.legend(['High Danger', 'Moderate Risk', 'Safe Zone'], loc='upper right')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
