```python
def impute_with_median(df):
    """Imputes missing values (NaN) in each column with the column's median"""
    for col in df.columns:
        if df[col].isnull().any():  # Check if the column has NaNs
            df[col].fillna(df[col].median(), inplace=True)
    return df 
df = impute_with_median(df) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the 'G2year' and 'G2fco2' columns

# Group by 'G2year' and calculate the average CO2 fugacity value for each year
average_fco2_by_year = df.groupby('G2year')['G2phtsinsitutp'].mean()

# Sort the average CO2 fugacity values in descending order
sorted_values = average_fco2_by_year.sort_values(ascending=False)

# Get the top two years with the highest average CO2 fugacity values
top_two_years = sorted_values.head(2)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
bar_plot = average_fco2_by_year.plot(kind='bar', color='#2F5233')
plt.title('Average Total Dissolved Inorganic Carbon (tco2) for Each Year')
plt.xlabel('Year')
plt.ylabel('Average tco2 Value')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding annotations only for the top two bars
for index, value in enumerate(average_fco2_by_year):
    if average_fco2_by_year.index[index] in top_two_years.index:
        plt.text(index, value + 0.05, f'{value:.2f}', ha='center')

plt.tight_layout()
plt.show()
```
